```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ltdaovn/dataset/master/ABC_Customers.csv')

df.describe()
```

```python
X = df.iloc[:, [3, 4]].values
# X
```

```python
clusters = []
for i in range(1, 10):
    model = KMeans(n_clusters=i).fit(X)
    clusters.append(model.inertia_)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(x=list(range(1, 10)), y=clusters, ax=ax)
ax.set_title('Đồ thị Elbow')
ax.set_xlabel('Số lượng nhóm')
ax.set_ylabel('Giá trị Inertia')
plt.show()
```

```python
# Qua đồ thị trên, chúng ta thấy số lượng cluster thích hợp là từ 3 đến 5 clusters

# Phân tích dữ liệu được gom thành 3 nhóm
model_3 = KMeans(n_clusters=3)
y_means = model_3.fit_predict(X)
print(y_means)

plt.scatter(X[y_means == 0, 0], X[y_means == 0, 1], s = 100, c = 'pink',
label = 'Nhóm 1')
plt.scatter(X[y_means == 1, 0], X[y_means == 1, 1], s = 100, c = 'yellow',
label = 'Nhóm 2')
plt.scatter(X[y_means == 2, 0], X[y_means == 2, 1], s = 100, c = 'cyan',
label = 'Nhóm 3')
plt.scatter(model_3.cluster_centers_[:,0], model_3.cluster_centers_[:, 1], s = 100, c = 'blue' , label = 'Centeroid')
plt.style.use('fivethirtyeight')
plt.title('K Means Clustering', fontsize = 20)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.grid()
plt.show()
```

```python
# Phân tích dữ liệu được gom thành 5 nhóm
model_5 = KMeans(n_clusters=5)
y_means = model_5.fit_predict(X)
print(y_means)

plt.scatter(X[y_means == 0, 0], X[y_means == 0, 1], s = 100, c = 'pink', label = 'Nhóm 1')
plt.scatter(X[y_means == 1, 0], X[y_means == 1, 1], s = 100, c = 'yellow', label = 'Nhóm 2')
plt.scatter(X[y_means == 2, 0], X[y_means == 2, 1], s = 100, c = 'cyan', label = 'Nhóm 3')
plt.scatter(X[y_means == 3, 0], X[y_means == 3, 1], s = 100, c = 'purple', label = 'Nhóm 4')
plt.scatter(X[y_means == 4, 0], X[y_means == 4, 1], s = 100, c = 'red', label = 'Nhóm 5')
plt.scatter(model_5.cluster_centers_[:,0], model_5.cluster_centers_[:, 1], s = 100, c = 'blue' , label = 'Centeroid')
plt.style.use('fivethirtyeight')
plt.title('K Means Clustering', fontsize = 20)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.grid()
plt.show()
```

- Nhóm 1 (Hồng):
    - Thu nhập hàng năm: trung bình, nằm trong khoảng 40-60.
    - Điểm chi tiêu: vừa phải, nằm trong khoảng 40-60.

- Nhóm 2 (Vàng):
    - Thu nhập hàng năm: khá cao, nằm trong khoảng 80-130.
    - Điểm chi tiêu: tương đối thấp, nằm trong khoảng 0-40.

- Nhóm 3 (Xanh):
    - Thu nhập hàng năm: thấp, nằm trong khoảng 0-40.
    - Điểm chi tiêu: thấp, nằm trong khoảng 0-40.

- Nhóm 4 (Tím):
    - Thu nhập hàng năm: cao, nằm trong khoảng 70-120.
    - Điểm chi tiêu: cao, nằm trong khoảng 70-100.

- Nhóm 5 (Đỏ):
    - Thu nhập hàng năm: thấp, nằm trong khoảng 0-40.
    - Điểm chi tiêu: cao, nằm trong khoảng 60-100.

```python
X1 = df.iloc[:, [2, 4]].values

clusters = []
for i in range(1, 10):
    model = KMeans(n_clusters=i).fit(X1)
    clusters.append(model.inertia_)

fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(x=list(range(1, 10)), y=clusters, ax=ax)
ax.set_title('Đồ thị Elbow')
ax.set_xlabel('Số lượng nhóm')
ax.set_ylabel('Giá trị Inertia')
plt.show()
```

```python
# Phân tích dữ liệu được gom thành 3 nhóm
model_3_age = KMeans(n_clusters=3)
y_means = model_3_age.fit_predict(X1)
print(y_means)

plt.scatter(X1[y_means == 0, 0], X1[y_means == 0, 1], s = 100, c = 'pink',
label = 'Nhóm 1')
plt.scatter(X1[y_means == 1, 0], X1[y_means == 1, 1], s = 100, c = 'yellow',
label = 'Nhóm 2')
plt.scatter(X1[y_means == 2, 0], X1[y_means == 2, 1], s = 100, c = 'cyan',
label = 'Nhóm 3')
plt.scatter(model_3_age.cluster_centers_[:,0], model_3_age.cluster_centers_[:, 1], s = 100, c = 'blue' , label = 'Centeroid')
plt.style.use('fivethirtyeight')
plt.title('K Means Clustering', fontsize = 20)
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.legend()
plt.grid()
plt.show()
```

- Nhóm 1 (Hồng):
    - Độ tuổi: trải dài từ 20-70
    - Điểm chi tiêu: vừa phải, nằm trong khoảng 40-60.
    - Đa số người có mức chi tiêu trung bình trải dài ở các độ tuổi.

- Nhóm 2 (Vàng):
    - Độ tuổi: trải dài từ 20-70, tập trung nhiều trong khoảng 30-60
    - Điểm chi tiêu: thấp, nằm trong khoảng 0-30.
    - Nhóm này có thể là những người tiết kiệm hoặc ít có nhu cầu chi tiêu.

- Nhóm 3 (Xanh):
    - Độ tuổi: từ 20-40
    - Điểm chi tiêu: cao, nằm trong khoảng 60-100.
    - Nhóm này có xu hướng chi tiêu nhiều, có thể là nhóm khách hàng tiềm năng trong tiêu dùng cao cấp.

```python
X2 = df.iloc[:, [2, 3, 4]].values

clusters = []
for i in range(1, 10):
    model = KMeans(n_clusters=i).fit(X2)
    clusters.append(model.inertia_)

fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(x=list(range(1, 10)), y=clusters, ax=ax)
ax.set_title('Đồ thị Elbow')
ax.set_xlabel('Số lượng nhóm')
ax.set_ylabel('Giá trị Inertia')
plt.show()
```
