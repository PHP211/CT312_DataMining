```python
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
data = {
    'Sinh viên': ['SV1', 'SV2', 'SV3', 'SV4', 'SV5', 'SV6', 'SV7', 'SV8'],
    'Văn': [7, 4, 2, 0, 9, 6, 2, 8],
    'Toán': [3, 5, 4, 1, 7, 8, 2, 8]
}


# Create a DataFrame with index for the subjects
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
X = df[['Văn', 'Toán']].values
```

```python
model = linkage(X, method='single')
```

```python
plt.figure(figsize=(10, 5))
dendrogram(model, labels=df['Sinh viên'].values)
plt.title('Single-link Hierarchical Clustering Dendrogram')
plt.xlabel('Sinh viên')
plt.ylabel('Distance')
plt.show()
```
