```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = {
    'cont': [1.0, 1.5, 2.0, 3.0, 4.0, 4.5,  5.0,  5.5, 6.0, 6.5, 7.0, 8.0, 9.0, 10.0, 11.0, 12.0, 13.0, 14.0, 15.0],
    'strength': [6.3, 11.1, 20.0, 24.0, 26.1, 30.0, 33.8, 34.0, 38.1, 39.9, 42.0, 46.1, 53.1, 52.0, 52.5, 48.0, 42.8, 27.8, 21.9] 
}

data = pd.DataFrame(data)
# data

# Liên hệ giữa Age và Chol
plt.scatter(data['cont'], data['strength'], color='blue')
plt.xlabel('Content')
plt.ylabel('Strength')
plt.show()
```

```python
X = data[['cont']]
y = data['strength']
# print(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
# print(len(X_test))

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Coefficients: ", model.coef_)
print("Intercept: ", round(model.intercept_,4))
print("Độ căng =", round(model.intercept_,4), "+", round(model.coef_[0],4), "× Hàm lượng")

print("mean_squared_error: ", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
