```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = {
    'age': [46, 20, 52, 30, 57, 25, 28, 36, 22, 43, 57, 33, 22, 63, 40, 48, 28, 49],
    'bmi': [25.4, 20.6, 26.2, 22.6, 25.4, 23.1, 22.7, 24.9, 19.8, 25.3, 23.2, 21.8, 20.9, 26.7, 26.4, 21.2, 21.2, 22.8],
    'chol': [3.5, 1.9, 4.0, 2.6, 4.5, 3.0, 2.9, 3.8, 2.1, 3.8, 4.1, 3.0, 2.5, 4.6, 3.2, 4.2, 2.3, 4.0] 
}

data = pd.DataFrame(data)
# data

# Liên hệ giữa Age và Chol
plt.scatter(data['age'], data['chol'], color='blue')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

```python
# Xây dựng mô hình dự đoán Chol dựa vào Age
X = data[['age']]
y = data['chol']
# print(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
# print(len(X_test))

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Coefficients: ", model.coef_)
print("Intercept: ", round(model.intercept_,4))
print("Cholesterol =", round(model.intercept_,4), "+", round(model.coef_[0],4), "× Tuổi")

print("mean_squared_error: ", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# Liên hệ giữa cả age, bmi và chol
sns.pairplot(data, x_vars=['age', 'bmi'], y_vars='chol', height=4, aspect=1, kind='scatter')
plt.show()
sns.heatmap(data.corr(), cmap="YlGnBu", annot = True)
plt.show()
```

```python
# Xây dựng mô hình
X1 = data[['age', 'bmi']]
y1 = data['chol']

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.3, random_state = 20)
# print(X1_train)

model1 = LinearRegression()
model1.fit(X1_train, y1_train)

y1_pred = model1.predict(X1_test)

print("Coefficients: ", model1.coef_)
print("Intercept: ", round(model1.intercept_,4))
print("Cholesterol =", round(model1.intercept_,4), "+", round(model1.coef_[0],4), "× Tuổi", "+", round(model1.coef_[1],4), "× BMI")

print("mean_squared_error: ", np.sqrt(metrics.mean_squared_error(y1_test, y1_pred)))
```
