```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
X = np.array([[150, 147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
y = np.array([[90, 49, 50, 51, 54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T

# Xây dựng biểu đồ
plt.plot(X, y, 'ro', color="blue")
plt.axis([140, 190, 45, 100])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Xây dựng mô hình
model = linear_model.LinearRegression()
model.fit(X, y)

X1 = np.array([[145,185]]).T
y1 = model.predict(X1)
plt.plot(X, y, 'ro', color="blue")
plt.plot(X1, y1, color="violet")

plt.axis([140, 190, 45, 100])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Nhận xét:
- Tập dữ liệu có 1 phần tử ngoại lệ, nằm cách xa các phần tử còn lại nên đường hồi quy sẽ bị lệch so với sự phân bổ của các phần tử, kết quả này là không tốt.

Phương án xử lý:
- Dễ dàng nhất là loại bỏ điểm ngoại lệ này ra khỏi tập dữ liệu. Khi đó sẽ có kết quả chính xác.
